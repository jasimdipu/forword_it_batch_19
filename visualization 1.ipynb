```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.head(100).plot(figsize=(15, 10))
```

```python
iris.head(10).plot(kind='bar', figsize=(15, 5))
plt.xlabel("Number of flowers")
plt.ylabel("Height of the flowers")
plt.show()
```
